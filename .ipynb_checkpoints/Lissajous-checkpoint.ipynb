```python
import numpy as np
import matplotlib.pyplot as plt

A=1
nx=[1,2,3,4,5]
ny=[1,2,3,4,5]
O=linspace(0,2*np.pi)
d= [0, np.pi/4, np.pi/2]
nsides=5
k=0
while k < len(d):
    fig, axis=plt.subplots(nsides,nsides,figsize=(3,3),constrained_layout=True,)
    i=0
    while i < len(nx):
        j=0
        while j < len(ny):
            if nx[i] < ny[j]:
                x=A*np.sin(nx[i]*O)
                y=A*np.sin(ny[j]*O+d[k])
                axis[i][j].plot(x,y)
            axis[i][j].axis("off")
            j=j+1
        i=i+1
    k=k+1
plt.title('Figure 1: Figuras de Lissajous para los desfases: δ = [0, π/4, π/2] respectivamente')
plt.show()
```
