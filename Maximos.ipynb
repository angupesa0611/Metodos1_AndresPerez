```python
import requests as rq
URL = "https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EstrellaEspectro.txt"
data= rq.get(URL)
open("data.txt","wb").write(data.content)
with open("data.txt","r") as data:
    matriz=[]
    for line in data:
        linea=[]
        for word in line.split():
            linea.append(word)
        matriz.append(linea)
```

Se crea e imprime la matriz con los maximos de la función

```python
maximos=[]
for i in matriz:
    if(matriz.index(i) > 0 and len(matriz) > matriz.index(i)+1 and i[1] > matriz[matriz.index(i)+1][1] and i[1] > matriz[matriz.index(i)-1][1]):
        maximos.append(i)
print(maximos)
```

Se imprimen los datos de una manera más organizada

```python
for i in maximos:
    print(i)
print("la cantidad totas de maximos es :"+str(len(maximos)))
```
