```python
import numpy as np
import matplotlib.pyplot as plt
O=np.linspace(0,2*np.pi)
a=0
b=1
r=a+b*O
datos=[]
for i in O:
    datos.append(a+b*O)
print(datos)
```

```python
fig=plt.figure()
dibujo=fig.add_subplot(111, projection="polar")
dibujo.plot(O,r)
plt.show()
```
