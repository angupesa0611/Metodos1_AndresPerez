$$f(x+2h) = f(x) + 2hf'(x) + 2h^2f''(x) + \frac{4}{3}h^3f'''(x) + \frac{2}{3}h^4f^{(IV)}(x)$$
$$f(x-2h) = f(x) - 2hf'(x) + 2h^2f''(x) - \frac{4}{3}h^3f'''(x) + \frac{2}{3}h^4f^{(IV)}(x)$$
Se suman ambas ecuaciones y se obtiene:
$$f(x+2h) + f(x-2h) = 2f(x) + 4h^2f''(x) + \frac{4}{3}h^4f^{(IV)}(x)$$
Se despeja para $f''(x)$
$$f''(x)= \frac{f(x+2h) - 2f(x) + f(x-2h)}{4h^2} - \frac{1}{3}h^2f^{(IV)}(x)$$

```python
h=0.05
minx=-10
maxx=10
x=np.arange(minx, maxx+h, h)
def f(pnum):
    res=1/np.sqrt(1+np.exp(-(pnum**2)))
    return(res)
dosderivada=(f(x+2*h)-2*f(x)+f(x-2*h))/(4*h**2)
print(dosderivada)
```

$$f(x+h) = f(x) + hf'(x) + \frac{1}{2}h^2f''(x) + \frac{1}{6}h^3f'''(x) + \frac{1}{24}h^4f^{(IV)}(x)$$
$$f(x-h) = f(x) - hf'(x) + \frac{1}{2}h^2f''(x) - \frac{1}{6}h^3f'''(x) + \frac{1}{24}h^4f^{(IV)}(x)$$

$$f(x+h) + f(x-h) = 2f(x) + h^2f''(x) + \frac{1}{12}h^4f^{(IV)}(x)$$
Se despeja para $f''(x)$
$$f^{(IV)}(x)= 12\frac{f(x+h) + f(x-h) - 2f(x) - h^2f''(x)}{h^4} $$
