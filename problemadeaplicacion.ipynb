```python
import numpy as np
import matplotlib.pyplot as plt
```

a.

```python
minx=-4
maxx=4
cantejex=25
dx=(maxx-minx)/(cantejex-1)
x=np.arange(minx, maxx+dx, dx)

miny=-4
maxy=4
cantejey=25
dy=(maxy-miny)/(cantejey-1)
y=np.arange(miny, maxy+dy, dy)

X,Y = np.meshgrid(x,y)
```

b.

```python
def f(x,y):
    R=2
    V=2
    return (V*x*(1-(R**2)/(x**2+y**2)))  
```

c.

```python
def df(x,y):
    h=0.001
    R=2
    dx=np.zeros(x.shape)
    dy=np.zeros(x.shape)
    i=0
    while i < x.shape[0]:
        j=0
        while j < x.shape[1]:
            if x[i][j]**2+y[i][j]**2>=R**2:
                dx[i][j]=(f(x[i][j]+h,y[i][j])-f(x[i][j]-h,y[i][j]))/(2*h)
                dy[i][j]=(f(x[i][j],y[i][j]+h)-f(x[i][j],y[i][j]-h))/(2*h)
            j=j+1
        i=i+1
    return([dx,dy])
derivadas=df(X,Y)
Vx=derivadas[0]
Vy=derivadas[1]
```

d.

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,2,1)

for i in range(cantejex):
    for j in range(cantejey):
        ax.quiver(x[i],y[j],Vx[i,j],Vy[i,j],color='b',alpha=0.7)
```
